```python
import numpy as np
import matplotlib.pyplot as plt

def plot_acceleration_data_grid(acceleration_data_list):
    time = np.arange(len(acceleration_data_list[0])) / 4000.0  # Assuming 4000 Hz sampling rate

    num_datasets = len(acceleration_data_list)
    num_plots_per_row = 3
    num_rows = (num_datasets + num_plots_per_row - 1) // num_plots_per_row

    plt.figure(figsize=(15, 5 * num_rows))

    for i, acceleration_data in enumerate(acceleration_data_list):
        plt.subplot(num_rows, num_plots_per_row, i + 1)
        plt.plot(time, acceleration_data[:, 0], label='X')
        plt.plot(time, acceleration_data[:, 1], label='Y')
        plt.plot(time, acceleration_data[:, 2], label='Z')
        plt.title(f'Dataset {i + 1}')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Acceleration')
        plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
num_datasets = 9*3
acceleration_data_list = [np.random.randn(40000, 3) for _ in range(num_datasets)]

plot_acceleration_data_grid(acceleration_data_list)
```
